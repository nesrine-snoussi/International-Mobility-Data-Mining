```python
import csv
import random

# define a list of countries where the universities are situated
countries = ['United States', 'Canada', 'United Kingdom', 'Australia', 'Germany', 'France', 'Spain', 'China', 'Japan', 'South Korea']

# define the number of rows to generate in the dataset
num_rows = 1000

# define the probability of a scholarship offer (10% in this example)
scholarship_probability = 0.1

# define a list of computer science options that could be studied later
computer_science_options = ['Artificial Intelligence', 'Data Science', 'Cybersecurity', 'Machine Learning', 'Web Development', 'Computer Networks']

# generate rows with randomly filled data
rows = []
for i in range(num_rows):
    # generate a random offer title
    offer_title = f"International Mobility Program {i}"
        
    # generate a random university name
    offered_by = f"University of {random.choice(countries)} {i}"
        
    # extract the country from the university name
    country = offered_by.split('University of ')[1].split(' ')[0]
        
    # generate a random offer description
    offer_description = f"This program offers an opportunity for international students to study at {offered_by} in {country}."
        
    # generate a random offer type (masters, exchange semester, or double diploma)
    offer_type = random.choice(['Masters', 'Exchange Semester', 'Double Diploma'])
        
    # generate a random offer amount based on the offer type (50% chance of no scholarship)
    if offer_type == 'Masters':
        offer_amount = random.choice([f"${random.randint(10000, 20000)}", 'None'] if random.random() > scholarship_probability else [f"${random.randint(20000, 30000)}"])
    elif offer_type == 'Exchange Semester':
        offer_amount = random.choice([f"${random.randint(500, 1000)}", 'None'] if random.random() > scholarship_probability else [f"${random.randint(1000, 2000)}"])
    else: # Double Diploma
        offer_amount = random.choice([f"${random.randint(15000, 25000)}", 'None'] if random.random() > scholarship_probability else [f"${random.randint(25000, 35000)}"])
        
    # generate a random application deadline
    application_deadline = f"{random.randint(1, 30)} {random.choice(['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'])}, 2023"
        
    # generate a random eligibility criteria based on the offer type
    if offer_type == 'Masters':
        minimum_gpa = round(random.uniform(10.0, 19.90), 1)
        eligibility_criteria = f"Enrolled in a full-time {offer_type} program outside of {country} with a minimum GPA of {minimum_gpa}."
        chosen_option = random.choice(computer_science_options)
        eligibility_criteria += f" Must be pursuing a {chosen_option} option in Computer Science."
    else:
        eligibility_criteria = f"Enrolled in a {offer_type} program outside of {country}."
    
    # generate a random success level for the last 5 years (between 1 and 10)
    success_level = random.randint(1, 10)
    

# add the row to the list of rows
    rows.append([offer_title, offered_by, country, offer_description, offer_type, offer_amount, application_deadline, eligibility_criteria, success_level])
    # Write the dataset to a CSV file
with open('universities.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Offer Title', 'Offered By', 'Country', 'Offer Description', 'Offer Type', 'Offer Amount', 'Application Deadline', 'Eligibility Criteria', 'Option to Study Later', 'success_level'])
    for row in rows:
        writer.writerow(row)
print(f"{num_rows} rows have been generated and written to mobility_offers.csv.")
```
